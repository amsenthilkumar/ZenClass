```python
import nltk
import urllib
import bs4 as bs
import re
from nltk.corpus import stopwords
```

```python
# to get data from the source
source= urllib.request.urlopen("https://en.wikipedia.org/wiki/European_colonization_of_the_Americas").read()
```

```python
#parsing the data and creating the beautiful soup obj
soup= bs.BeautifulSoup(source,'lxml')
```

```python
# fetching the data
text= ""
for paragraph in soup.find_all('p'):
    text= text+paragraph.text
```

```python
text
```

```python
# data preprocessing
text = re.sub(r'\[[0-9]*\]',' ',text)
```

```python
text
```

```python
# decorating the text(removing the double spaces, lowering the case and removing digits from strings)
text = re.sub(r'\s+',' ',text)
text= text.lower()
text= re.sub(r'\d','',text)
text = re.sub(r'\s+',' ',text)
```

```python
text
```

```python
import nltk
nltk.download('punkt')
sentences=nltk.sent_tokenize(text)
```

```python
sentences= [nltk.word_tokenize(sentence) for sentence in sentences ]
```
